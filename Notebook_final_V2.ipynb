# Prédiction de défaut des prêts SBA

---
# Résumé du Projet

Ce projet, construit sur les prêts historiques du programme **7(a) de la SBA** (US Small Business Administration), vise à prédire le risque de non remboursement (`LoanStatus`) d'une entreprise participant à ce programme. Ainsi l'objectif du modèle est de distinguer efficacement les prêts remboursés (*Paid in Full*) des pertes sèches (*Charged Off*). 

Le projet est construit en plusieurs étapes :

1.  **Analyse Temporelle :** Au début du projet, les données disponibles depuis le début des années 2000 ont été analysées afin de comprendre les relations entre les différentes variables disponibles ainsi que l'évolution dans le temps des défauts. L'analyse a montré la **nécessité** de réduire la période d'étude suite aux variations des cycles macroéconomiques observées dans le temps. Ainsi, la période d'étude a été réduite à **2020-2024** afin de capter les dernières tendances économiques et éviter d'apprendre sur des données qui ne sont aujourd'hui plus d'actualité (Idéalement nous aurions voulu prendre les données après 2022 où on a un pic de défaut anormale résultant des effets du covid, mais impossible au vu de la faible quantité de données sur cette période). 

2.  **Enrichissement Macroéconomique :** En plus des variables disponibles dans les tables fournies par la SBA, nous avons ajouté des **variables macroéconomiques** au modèle afin de lui donner au mieux des indications sur les périodes de crise et de changement macroéconomique. Afin d'éviter tout *data leakage* possible, mais aussi dans le but de construire un projet réaliste et non seulement maximiser nos métriques, ces variables macroéconomiques ont été utilisées avec un **décalage temporel** (ex : croissance du PIB de l'année précédente et non de l'année actuelle).

3.  **Découpage Train/Test :** Le découpage en échantillon *train/test* respecte la temporalité. Afin de valider les performances du modèle avant de le tester sur des données que le modèle n'aura jamais vu nous avons opté pour un time series split cross validation. 


5.  **Modélisation Hybride :** Plusieurs modèles seront comparés, en commençant par une **régression logistique (avec pénalisation et régularisation)** en tant que modèle benchmark, puis des modèles plus avancés (**Boosting, Bagging**) seront également testés. 

# Le Dataset (SBA 7(a) & 504)
Les données de départ (disponibles directement sur https://data.sba.gov/dataset/) contiennent plusieurs variables de différentes nature sur :
* **L'Emprunteur :** Localisation (`ProjectState`), Secteur d'activité (`NAICS`, `Industry`), Ancienneté.
* **La Structure du Prêt :** Montant (`GrossApproval`), Partie Garantie (`SBAGuaranteedApproval`), Durée (`TermInMonths`), Type (`RevolverStatus`).
* **Le Prêteur :** Qualité de la banque (`BankName`) et méthode d'octroi (`ProcessingMethod`).
* **Impact & Collatéral :** Emplois crées (`JobsSupported`) et présence de garanties (`CollateralInd`).

## 7.2. Modèle simple pour commencer (Elastic Net)

```python
cat_features = [
    'Subprogram', 'ProjectState', 'BusinessType',
    'RevolverStatus', 'CollateralInd', 'IndustryName', 'TerminMonths_cut', "JobsSupported_cut"
]

num_features = [col for col in X_train.columns if col not in cat_features]
```

```python
# On fait une fonction poru filtrer les variables trop corrlées pour une rég log
seuil_correlation = 0.6

def compute_corr_filter_indices(X_array, y_array, threshold=0.8):
    X_array = np.asarray(X_array)
    y_array = np.asarray(y_array).ravel().astype(float)

    n_features = X_array.shape[1]

    # On prend la corrélation avec la cible
    corr_with_y = np.zeros(n_features)
    for i in range(n_features):
        xi = X_array[:, i]
        if np.std(xi) == 0:
            corr_with_y[i] = 0.0
        else:
            corr_with_y[i] = np.corrcoef(xi, y_array)[0, 1]

    # Matrice de corrélation entre les variables
    corr_matrix = np.corrcoef(X_array, rowvar=False)
    corr_matrix = np.nan_to_num(corr_matrix)

    to_drop = np.zeros(n_features, dtype=bool)

    for i in range(n_features):
        if to_drop[i]:
            continue
        for j in range(i + 1, n_features):
            if to_drop[j]:
                continue
            if abs(corr_matrix[i, j]) > threshold:
                # Parmi deux var trp corrélées on prend celle qui est la plus corréléé à la cible
                if abs(corr_with_y[i]) < abs(corr_with_y[j]):
                    to_drop[i] = True
                    break
                else:
                    to_drop[j] = True

    keep_indices = np.where(~to_drop)[0]
    return keep_indices


def fit_preprocessors(X_tr, y_tr, corr_threshold=0.8):
    # Séparation num / cat
    X_tr_num = X_tr[num_features].values
    X_tr_cat = X_tr[cat_features].values

    # Standardisation
    scaler = StandardScaler()
    X_tr_num_scaled = scaler.fit_transform(X_tr_num)

    # Encodage catégoriel
    encoder = OneHotEncoder(
        handle_unknown="infrequent_if_exist",
        min_frequency=0.01,
        sparse_output=False
    )
    X_tr_cat_enc = encoder.fit_transform(X_tr_cat)

    # concat num cat 
    X_tr_proc = np.hstack([X_tr_num_scaled, X_tr_cat_enc])
    cat_feature_names = encoder.get_feature_names_out(cat_features)
    feature_names = np.concatenate([num_features, cat_feature_names])
    
    stds = np.std(X_tr_proc, axis=0)
    non_const_indices = np.where(stds > 1e-9)[0] 
    

    
    X_tr_proc_active = X_tr_proc[:, non_const_indices]


    
    keep_indices_local = compute_corr_filter_indices(X_tr_proc_active, y_tr, threshold=seuil_correlation)

    keep_indices = non_const_indices[keep_indices_local]

    X_tr_filt = X_tr_proc[:, keep_indices]
    feature_names_filtered = feature_names[keep_indices]

    return X_tr_filt, scaler, encoder, keep_indices, feature_names_filtered

# On utilise les éléments de la fonction préc appris sur le train du fold
def transform_with_preprocessors(X, scaler, encoder, keep_indices):
    X_num = X[num_features].values
    X_cat = X[cat_features].values

    X_num_scaled = scaler.transform(X_num)
    X_cat_enc = encoder.transform(X_cat)

    X_proc = np.hstack([X_num_scaled, X_cat_enc])
    X_filt = X_proc[:, keep_indices]

    return X_filt
```

```python
tscv = TimeSeriesSplit(n_splits=5)
cv_aucs = []

fold = 1
for train_idx, val_idx in tscv.split(X_train):
    X_tr_fold = X_train.iloc[train_idx]
    y_tr_fold = y_train.iloc[train_idx]

    X_val_fold = X_train.iloc[val_idx]
    y_val_fold = y_train.iloc[val_idx]

    X_tr_filt, scaler_fold, encoder_fold, keep_indices_fold, feature_names_fold = fit_preprocessors(
    X_tr_fold, y_tr_fold, seuil_correlation)


    X_val_filt = transform_with_preprocessors(
        X_val_fold, scaler_fold, encoder_fold, keep_indices_fold
    )

    logreg_en = LogisticRegression(
        penalty="elasticnet",
        solver="saga",
        l1_ratio=0.5,  # Pour l'instant on fixe un truc manuellement        
        C=1.0,
        max_iter=5000,
        class_weight="balanced", #Pour gérer le déséqulibre des classes
        random_state=42
    )

    # Fit sur le train du fold
    logreg_en.fit(X_tr_filt, y_tr_fold)

    # perf sur le validation du fold
    y_val_proba = logreg_en.predict_proba(X_val_filt)[:, 1]
    auc_fold = roc_auc_score(y_val_fold, y_val_proba)
    cv_aucs.append(auc_fold)

    print(f"Fold {fold} - AUC: {auc_fold:.4f}")
    fold += 1

print("\nAUC moyenne (CV) :", np.mean(cv_aucs))
print("Écart-type AUC   :", np.std(cv_aucs))
```

```python
X_train_filt, scaler_full, encoder_full, keep_indices_full, feature_names_filtered = fit_preprocessors(
    X_train, y_train, seuil_correlation
)

X_test_filt = transform_with_preprocessors(
    X_test, scaler_full, encoder_full, keep_indices_full
)

# Modèle final
logreg_final = LogisticRegression(
    penalty="elasticnet",
    solver="saga",
    l1_ratio=0.5, # on fixe des param simples pour l'instant
    C=1.0,
    max_iter=5000,
    n_jobs=-1,
    class_weight="balanced",
    random_state=42
)

logreg_final.fit(X_train_filt, y_train)

# AUC sur le TEST
y_test_proba = logreg_final.predict_proba(X_test_filt)[:, 1]
y_test_pred  = logreg_final.predict(X_test_filt)

test_auc = roc_auc_score(y_test, y_test_proba)
print("\nAUC sur le jeu de TEST :", test_auc)
```

```python
coefs = logreg_final.coef_.ravel()
importance = pd.DataFrame({
    "feature": feature_names_filtered,
    "coef": coefs,
    "abs_coef": np.abs(coefs)
}).sort_values("abs_coef", ascending=False)

print("\nvariables les plus importantes :")
print(importance.head(10))
```

```python
tscv = TimeSeriesSplit(n_splits=5)
seuil_correlation = 0.6
    
# ##############################################################################
# 1. FILTRAGE DE CORRÉLATIONS (Spearman)
# ##############################################################################

print("*" * 100)
print(f"--- 1. FILTRAGE DES CORRÉLATIONS (Spearman > {seuil_correlation}) ---")
print("----- ON garde les variables les plus corrélées à y ----")
print("*" * 100)

feature_names = X_train_encoded.columns.tolist()

corr_matrix_X = X_train_encoded.corr(method="spearman").abs()
corr_X_Y = X_train_encoded.apply(lambda x: x.corr(y_train)).abs().sort_values(ascending=False)

vars_to_check = corr_X_Y.index.tolist()
vars_to_drop = set()
vars_uncorrelated = []

for i in range(len(vars_to_check)):
    col_1 = vars_to_check[i]
    if col_1 in vars_to_drop:
        continue
    
    vars_uncorrelated.append(col_1)
    
    for j in range(i + 1, len(vars_to_check)):
        col_2 = vars_to_check[j]
        if col_2 in vars_to_drop:
            continue
        if corr_matrix_X.loc[col_1, col_2] > seuil_correlation:
            vars_to_drop.add(col_2)

print(f"\nVariables de départ : {len(feature_names):,}")
print(f"Variables ayant passé le filtre : {len(vars_uncorrelated):,}")

X_train_filtered = X_train_encoded[vars_uncorrelated].copy()
X_test_filtered  = X_test_encoded[vars_uncorrelated].copy()


# ##############################################################################
# 2. SÉLECTION DE VARIABLES PAR LASSO (L1)
# ##############################################################################

scaler_lasso = StandardScaler()
X_train_scaled_lasso = scaler_lasso.fit_transform(X_train_filtered)
X_train_scaled_lasso = pd.DataFrame(
    X_train_scaled_lasso,
    columns=X_train_filtered.columns,
    index=X_train_filtered.index
)

lasso_model = LogisticRegression(
    penalty='l1',
    solver='saga',
    max_iter=5000,
    random_state=42,
    class_weight='balanced'
)

param_grid_lasso = {'C': [0.001, 0.01, 0.05, 0.1, 0.5, 1.0]}

grid_search_lasso = GridSearchCV(
    lasso_model,
    param_grid_lasso,
    cv=tscv,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=0
)

grid_search_lasso.fit(X_train_scaled_lasso, y_train)

best_lasso = grid_search_lasso.best_estimator_
best_C = best_lasso.C

print(f"\nMeilleur paramètre Lasso : C = {best_C}")

coeffs = best_lasso.coef_[0]
feature_names_filtered = X_train_filtered.columns

lasso_coefs = pd.DataFrame({
    'Variable': feature_names_filtered,
    'Coefficient': coeffs,
    'Importance_Absolue': np.abs(coeffs)
})

vars_final = lasso_coefs[lasso_coefs['Coefficient'] != 0]['Variable'].tolist()

print(f"Variables sélectionnées par Lasso : {len(vars_final)}")

X_train_final = X_train_filtered[vars_final].copy()
X_test_final  = X_test_filtered[vars_final].copy()


# ##############################################################################
# 3. LOGISTIC REGRESSION (L2) AVEC TIME SERIES SPLIT
# ##############################################################################

logreg_base = LogisticRegression(
    penalty='l2', 
    solver='lbfgs',
    max_iter=5000,
    random_state=42,
    class_weight='balanced'
)

param_grid_logreg = {'C': [0.01, 0.1, 1.0, 10.0]}

auc_scores_logreg = []
best_params_logreg_per_fold = []

print("\n--- AUC PAR FOLD (TimeSeriesSplit) ---\n")

for fold, (train_index, val_index) in enumerate(tscv.split(X_train_final), 1):
    
    X_train_fold, X_val_fold = X_train_final.iloc[train_index], X_train_final.iloc[val_index]
    y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]
    
    scaler_fold = StandardScaler()
    X_train_scaled = scaler_fold.fit_transform(X_train_fold)
    X_val_scaled   = scaler_fold.transform(X_val_fold)
    
    inner_tscv = TimeSeriesSplit(n_splits=3)
    
    inner_grid_search = GridSearchCV(
        logreg_base,
        param_grid_logreg,
        scoring='roc_auc',
        cv=inner_tscv,
        verbose=0,
        n_jobs=-1
    )
    
    inner_grid_search.fit(X_train_scaled, y_train_fold)
    
    best_logreg_fold = inner_grid_search.best_estimator_
    best_C_fold = inner_grid_search.best_params_['C']
    best_params_logreg_per_fold.append(best_C_fold)
    
    probs_val_fold = best_logreg_fold.predict_proba(X_val_scaled)[:, 1]
    auc_fold = roc_auc_score(y_val_fold, probs_val_fold)
    auc_scores_logreg.append(auc_fold)
    
    print(f"Fold {fold} : AUC = {auc_fold:.4f}   (Meilleur C = {best_C_fold})")

auc_cv_mean = np.mean(auc_scores_logreg)
optimal_C   = np.mean(best_params_logreg_per_fold)

print(f"\nAUC MOYENNE (CV) : {auc_cv_mean:.4f}")
print(f"PARAMÈTRE C FINAL = {optimal_C:.4f}")


# ##############################################################################
# 4. ENTRAÎNEMENT FINAL + AUC TEST
# ##############################################################################

final_scaler = StandardScaler()
X_train_final_scaled = final_scaler.fit_transform(X_train_final)
X_test_final_scaled  = final_scaler.transform(X_test_final)

logreg_final_model = LogisticRegression(
    penalty='l2',
    C=optimal_C, 
    solver='lbfgs',
    max_iter=5000,
    random_state=42,
    class_weight='balanced'
)

logreg_final_model.fit(X_train_final_scaled, y_train)

probs_test = logreg_final_model.predict_proba(X_test_final_scaled)[:, 1]
auc_test = roc_auc_score(y_test, probs_test)

print(f"\n>>> AUC FINAL SUR LE TEST : {auc_test:.4f}")
```

## 7.7. XGBOOST 

```python
tscv = TimeSeriesSplit(n_splits=5)
seuil_correlation = 1 # On prend finalement tout après plusieurs essais

print("*" * 100)
print(f"--- 1. FILTRAGE DES CORRÉLATIONS (Spearman > {seuil_correlation}) ---")
print("----- ON garde les variables les plus corrélées à y ----")
print("*" * 100)

feature_names = X_train_encoded.columns.tolist()
corr_matrix_X = X_train_encoded.corr(method="spearman").abs()

# Corrélation absolue de chaque feature avec la cible
corr_X_Y = X_train_encoded.apply(
    lambda x: x.corr(y_train, method="spearman")
).abs().sort_values(ascending=False)

vars_to_check = corr_X_Y.index.tolist()
vars_to_drop = set()
vars_uncorrelated = []

for i in range(len(vars_to_check)):
    col_1 = vars_to_check[i]
    if col_1 in vars_to_drop:
        continue
    
    # On garde col_1
    vars_uncorrelated.append(col_1)
    
    # On supprime les variables trop corrélées à col_1
    for j in range(i + 1, len(vars_to_check)):
        col_2 = vars_to_check[j]
        if col_2 in vars_to_drop:
            continue
        if corr_matrix_X.loc[col_1, col_2] > seuil_correlation:
            vars_to_drop.add(col_2)

print(f"\nVariables de départ : {len(feature_names):,}")
print(f"Variables supprimées (corrélation > {seuil_correlation}) : {len(vars_to_drop)}")
print(f"Variables non corrélées retenues : {len(vars_uncorrelated):,}")

# Dataset filtré (base pour XGBoost)
X_train_filtered = X_train_encoded[vars_uncorrelated].copy()
X_test_filtered  = X_test_encoded[vars_uncorrelated].copy()

# =============================================================================
# 2. XGBOOST : DÉFINITION DU MODÈLE & GRILLE D'HYPERPARAMÈTRES
# =============================================================================

# Gestion du déséquilibre de classes
count_neg = np.sum(y_train == 0)
count_pos = np.sum(y_train == 1)
scale_pos_weight = np.sqrt(count_neg / count_pos)  # pondération simple

xgb_base = XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    random_state=42,
    n_jobs=-1,
    scale_pos_weight=scale_pos_weight,
    tree_method="hist" 
)

param_grid_xgb = {
    'n_estimators': [300, 400,],            
    'learning_rate': [0.01],        
    'max_depth': [4, 5],               
    'subsample': [0.6, 0.8],         
    'colsample_bytree': [0.6, 0.8, 1],  
    'reg_alpha': [1, 5],               
    'reg_lambda': [3, 6],              
    'min_child_weight': [1, 10]          
}

grid_search_xgb = GridSearchCV(
    estimator=xgb_base,
    param_grid=param_grid_xgb,
    scoring='roc_auc',
    cv=tscv,          
    verbose=1,
    n_jobs=-1
)

print("\n--- 2. GRIDSEARCH XGBOOST (CV temporelle) ---\n")
grid_search_xgb.fit(X_train_filtered, y_train)

best_xgb_params = grid_search_xgb.best_params_
auc_cv_mean = grid_search_xgb.best_score_

print("\nMeilleurs hyperparamètres XGBoost (GridSearchCV) :")
for k, v in best_xgb_params.items():
    print(f" - {k}: {v}")
print(f"\nAUC Moyenne Validation (CV - GridSearch) : {auc_cv_mean:.4f}")


auc_scores_folds = []

for fold, (train_index, val_index) in enumerate(tscv.split(X_train_filtered), 1):
    X_train_fold, X_val_fold = X_train_filtered.iloc[train_index], X_train_filtered.iloc[val_index]
    y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]
    
    xgb_fold = XGBClassifier(
        objective='binary:logistic',
        eval_metric='auc',
        random_state=42,
        n_jobs=-1,
        scale_pos_weight=scale_pos_weight,
        tree_method="hist",
        **best_xgb_params
    )
    
    xgb_fold.fit(X_train_fold, y_train_fold)
    
    probs_val_fold = xgb_fold.predict_proba(X_val_fold)[:, 1]
    auc_fold = roc_auc_score(y_val_fold, probs_val_fold)
    auc_scores_folds.append(auc_fold)
    
    print(f"Fold {fold} : AUC = {auc_fold:.4f}")

auc_cv_mean_manual = np.mean(auc_scores_folds)
print(f"\nAUC MOYENNE (CV - recalculée avec best params) : {auc_cv_mean_manual:.4f}")

# =============================================================================
# 4. ENTRAÎNEMENT FINAL SUR TOUT LE TRAIN + AUC TEST
# =============================================================================

print("\n--- 4. MODÈLE FINAL XGBOOST + AUC TEST ---\n")

xgb_final_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    random_state=42,
    n_jobs=-1,
    scale_pos_weight=scale_pos_weight,
    tree_method="hist",
    **best_xgb_params
)

xgb_final_model.fit(X_train_filtered, y_train)

probs_test = xgb_final_model.predict_proba(X_test_filtered)[:, 1]
auc_test = roc_auc_score(y_test, probs_test)

print(f">>> AUC FINAL SUR LE TEST : {auc_test:.4f}")
```

## 7.8. CatBoost sans encodage préalable (on prend l'encodage natif de catboost) + OPTUNA

```python
categorical_features_list = ['Subprogram', 'BusinessType', 'IndustryName', 'ProjectState', "TerminMonths_cut", "JobsSupported_cut"]
tscv = TimeSeriesSplit(n_splits=5)
seuil_correlation = 1 #On prend tout finalement

print("*" * 100)
print(f"--- 1. FILTRAGE DES CORRÉLATIONS (Spearman > {seuil_correlation}) ---")
print("----- Uniquement sur les variables numériques ----")
print("*" * 100)

X_train_numeric = X_train.select_dtypes(include=np.number)
X_test_numeric  = X_test.select_dtypes(include=np.number)

numeric_cols = X_train_numeric.columns.tolist()
corr_matrix_X = X_train_numeric.corr(method="spearman").abs()

corr_X_Y = X_train_numeric.apply(
    lambda x: x.corr(y_train, method="spearman")
).abs().sort_values(ascending=False)

vars_to_check = corr_X_Y.index.tolist()
vars_to_drop = set()
vars_uncorrelated_numeric = []

for i in range(len(vars_to_check)):
    col_1 = vars_to_check[i]
    if col_1 in vars_to_drop:
        continue
    vars_uncorrelated_numeric.append(col_1)
    for j in range(i + 1, len(vars_to_check)):
        col_2 = vars_to_check[j]
        if col_2 in vars_to_drop:
            continue
        if corr_matrix_X.loc[col_1, col_2] > seuil_correlation:
            vars_to_drop.add(col_2)

vars_final = vars_uncorrelated_numeric + categorical_features_list

print(f"\nVariables Numériques de départ : {len(numeric_cols):,}")
print(f"Variables Numériques non corrélées retenues : {len(vars_uncorrelated_numeric):,}")
print(f"Variables Catégorielles incluses : {len(vars_final) - len(vars_uncorrelated_numeric):,}")

X_train_final = X_train[vars_final].copy()
X_test_final  = X_test[vars_final].copy()


for col in categorical_features_list:
    X_train_final[col] = X_train_final[col].astype(str)
    X_test_final[col]  = X_test_final[col].astype(str)


def objective(trial):

    params = {
        "loss_function": "Logloss",
        "eval_metric": "AUC",
        "random_seed": 42,
        "verbose": 0,
        "auto_class_weights": "Balanced",

        "learning_rate": trial.suggest_float("learning_rate", 0.005, 0.1, log=True),
        "n_estimators": trial.suggest_int("n_estimators", 200, 800),
        "depth": trial.suggest_int("depth", 3, 8),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1, 10),
        "bootstrap_type": "Bernoulli",
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
    }

    auc_scores = []

    for train_idx, val_idx in tscv.split(X_train_final):

        X_train_fold = X_train_final.iloc[train_idx]
        X_val_fold   = X_train_final.iloc[val_idx]
        y_train_fold = y_train.iloc[train_idx]
        y_val_fold   = y_train.iloc[val_idx]

        model = CatBoostClassifier(**params)

        model.fit(
            X_train_fold, y_train_fold,
            cat_features=categorical_features_list
        )

        preds = model.predict_proba(X_val_fold)[:, 1]
        auc = roc_auc_score(y_val_fold, preds)

        auc_scores.append(auc)

    return np.mean(auc_scores)


print("\n--- 2. OPTIMISATION OPTUNA (CV temporelle) ---\n")

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=40, show_progress_bar=True)

best_params = study.best_params
best_auc = study.best_value

print("\nMeilleurs hyperparamètres trouvés par Optuna :")
for k, v in best_params.items():
    print(f" - {k}: {v}")

print(f"\nAUC Moyenne Validation (CV - Optuna) : {best_auc:.4f}")


print("\n--- 3. AUC PAR FOLD (avec les meilleurs paramètres Optuna) ---\n")

auc_scores_folds = []

for fold, (train_idx, val_idx) in enumerate(tscv.split(X_train_final), 1):

    X_train_fold = X_train_final.iloc[train_idx]
    X_val_fold   = X_train_final.iloc[val_idx]
    y_train_fold = y_train.iloc[train_idx]
    y_val_fold   = y_train.iloc[val_idx]

    model = CatBoostClassifier(
        loss_function="Logloss",
        eval_metric="AUC",
        random_seed=42,
        verbose=0,
        auto_class_weights="Balanced",
        bootstrap_type="Bernoulli",
        **best_params
    )

    model.fit(
        X_train_fold, y_train_fold,
        cat_features=categorical_features_list
    )

    probs = model.predict_proba(X_val_fold)[:, 1]
    auc_fold = roc_auc_score(y_val_fold, probs)

    auc_scores_folds.append(auc_fold)
    print(f"Fold {fold} : AUC = {auc_fold:.4f}")

auc_mean_manual = np.mean(auc_scores_folds)
print(f"\nAUC MOYENNE (CV - recalculée avec best params) : {auc_mean_manual:.4f}")


print("\n--- 4. MODÈLE FINAL CATBOOST + AUC TEST ---\n")

cat_final_model = CatBoostClassifier(
    loss_function="Logloss",
    eval_metric="AUC",
    random_seed=42,
    verbose=0,
    auto_class_weights="Balanced",
    bootstrap_type="Bernoulli",
    **best_params
)

cat_final_model.fit(
    X_train_final, y_train,
    cat_features=categorical_features_list
)

probs_test = cat_final_model.predict_proba(X_test_final)[:, 1]
auc_test = roc_auc_score(y_test, probs_test)

print(f">>> AUC FINAL SUR LE TEST : {auc_test:.4f}")
```

```python
print("\n--- 5. IMPORTANCE DES VARIABLES (CatBoost) ---\n")

importance = cat_final_model.get_feature_importance()

importance_df = pd.DataFrame({
    'Variable': X_train_final.columns,
    'Importance': importance
})

importance_df = importance_df.sort_values(by='Importance', ascending=True)

plt.figure(figsize=(10, 12))
sns.set_style("whitegrid", {'grid.linestyle': '--'})

plt.barh(
    importance_df['Variable'],
    importance_df['Importance'],
    color=sns.color_palette("viridis", 1)[0]
)

plt.title("Importance des Variables (CatBoost)", fontsize=16, pad=20)
plt.xlabel("Importance (Score CatBoost)", fontsize=12)
plt.ylabel("Variable", fontsize=12)

for index, value in enumerate(importance_df['Importance']):
    plt.text(value + 0.0001, index, f"{value:.4f}", va='center')

plt.tight_layout()
plt.show()
```

## 8.1 Régression logistique LASSO

```python
lasso_model = LogisticRegression(
    penalty='l1', 
    solver='liblinear', 
    max_iter=2000, 
    random_state=42, 
    class_weight='balanced'
)

param_grid = {'C': [0.0001, 0.001, 0.005, 0.01, 0.05, 0.1]}

tscv = TimeSeriesSplit(n_splits=5)

grid_search = GridSearchCV(
    lasso_model, 
    param_grid, 
    cv=tscv, 
    scoring='roc_auc', 
    n_jobs=-1,
    verbose=0
)

grid_search.fit(X_train_encoded, y_train)
best_lasso = grid_search.best_estimator_
best_C = grid_search.best_params_['C']

coeffs = best_lasso.coef_[0]
feature_names = X_train_encoded.columns

lasso_coefs = pd.DataFrame({
    'Variable': feature_names,
    'Coefficient': coeffs,
    'Importance_Absolue': np.abs(coeffs)
})

survivors = lasso_coefs[lasso_coefs['Coefficient'] != 0].sort_values(by='Importance_Absolue', ascending=False)
vars_to_check = survivors['Variable'].tolist()

corr_matrix = X_train_encoded[vars_to_check].corr(method="spearman").abs()

vars_to_drop = set()
vars_final = []

for i in range(len(vars_to_check)):
    col_1 = vars_to_check[i]
    
    if col_1 in vars_to_drop:
        continue
    
    vars_final.append(col_1)
    
    for j in range(i + 1, len(vars_to_check)):
        col_2 = vars_to_check[j]
        
        if col_2 in vars_to_drop:
            continue
            
        if corr_matrix.loc[col_1, col_2] > 0.4:
            vars_to_drop.add(col_2)

X_train_final = X_train_encoded[vars_final]
X_test_final = X_test_encoded[vars_final]
X_oot_final = X_oot_encoded[vars_final]

final_model = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    C=best_C,
    max_iter=2000,
    random_state=42,
    class_weight='balanced'
)

final_model.fit(X_train_final, y_train)

probs_train = final_model.predict_proba(X_train_final)[:, 1]
probs_test = final_model.predict_proba(X_test_final)[:, 1]
probs_oot = final_model.predict_proba(X_oot_final)[:, 1]

auc_train = roc_auc_score(y_train, probs_train)
auc_test = roc_auc_score(y_test, probs_test)
auc_oot = roc_auc_score(y_oot, probs_oot)

print(f"Variables initiales : {len(feature_names)}")
print(f"Variables après Lasso : {len(survivors)}")
print(f"Variables finales (Lasso + Corr < 0.4) : {len(vars_final)}")
print("-" * 30)
print(f"AUC Train Final : {auc_train:.4f}")
print(f"AUC Test Final  : {auc_test:.4f}")
print(f"AUC OOT Final   : {auc_oot:.4f}")
```

```python
def calculate_gini(auc_score):
    return 2 * auc_score - 1


def relative_gini(gini_train, gini_test):
    if gini_train == 0:
        return 0
    return (gini_train - gini_test) / gini_train
```
